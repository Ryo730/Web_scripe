```python
##一つ目のデータを取得
def first_data(url,when):
    global data_sum
    data=pd.read_html(url,header=0)
    
    data=data[0]
    data=data[['国名',dece]]
    data.index=data['国名']
    data=data.drop('国名',axis=1)
    data.columns=[when] 
    data_sum=data
    return data_sum.head()

##二つ目以降は追加していく
def merge_data(url):
    global data1,data_sum
    data=pd.read_html(url,header=0)
    check=data_sum
    num=len(check)-1
    try:
        when_is=data_sum.columns[-1]
        print('Correct')
    except:
        when_is=data_sum.columns[0]
        print('False')
    if len(check.columns)%2==0:
        month=dece
        when_is=when_is-94
        print('12月')
    else :
        month=june
        when_is=when_is-6
        print("6月")
    print('when_is:{}'.format(when_is))
    data=data[0][['国名',month]]
    data.index=data['国名']
    data=data.drop('国名',axis=1)
    data.columns=[when_is]
    
    data1=data
    data_sum=data_sum.join([data1])
    return data_sum.head()
```

```python
##列タグのみのリスト作成
columns=[]
for i in data_sum.columns:
    columns.append(i)
```

```python
## '位'を取り除く
#各columnを入れて行くことしか対応していないため，for文で対応した．
for i in columns:
    data_sum[i]=data_sum[i].str.replace('位','')
```

```python
##　数字がobject型なので，floatへと変換した
#スライスは行指定でしか使えない
#NaNがあると，対応しないため９９に変更
data_sum[columns]=data_sum[columns].fillna(99)
data_sum[columns]=data_sum[columns].astype('int64')
```
